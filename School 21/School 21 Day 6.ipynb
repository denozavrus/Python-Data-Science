```python
%%timeit
new_column = []

for i in range(0, len(fines)): 
    value = fines.iloc[i, 2] / (fines.iloc[i, 1] * fines.iloc[i, 5])
    new_column.append ( value )
    
fines["New_column"] = new_column
```

```python
fines.head()
```

```python
fines.drop(["New_column"], axis = 1, inplace=True)
```

```python
%%timeit
new_column = []
for _, row in fines.iterrows(): 
    value = row.loc["Fines"] / (row.loc["Refund"] * row.loc["Year"])
    new_column.append(value)

fines["New_column"] = new_column++
```

```python
fines.drop(["New_column"], axis = 1, inplace=True)
```

```python
%%timeit
fines["New_column"] = fines.apply(lambda x: x.loc["Fines"] / (x.loc["Refund"] * x.loc["Year"]),
                                 axis = 1)
```
